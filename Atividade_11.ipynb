```python
import json
import numpy as np
!pip install faiss-cpu
import faiss
from sentence_transformers import SentenceTransformer

# carrega modelo
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# carrega índice FAISS
index = faiss.read_index("vector_store/faiss_index.faiss")

# carrega metadados
with open("vector_store/meta.json", "r", encoding="utf-8") as f:
    meta = json.load(f)

def query_text(q, top_k=5):
    # gera embedding para a query
    q_emb = model.encode([q], convert_to_numpy=True)
    faiss.normalize_L2(q_emb)
    # busca no índice
    D, I = index.search(q_emb, top_k)
    # D = distâncias, I = índices dos vetores similares
    results = []
    for dist, idx in zip(D[0], I[0]):
        results.append(
            {
                "score": float(dist),
                "text": meta[idx]["text"],
                "id": meta[idx]["id"],
            }
        )
    return results

if __name__ == "__main__":
    while True:
        q = input("Pergunta: ")
        res = query_text(q, top_k=3)
        for r in res:
            print(f"- (score: {r['score']:.4f}) {r['text']}")
```
